## Inheritence 

**Inheritance** allows us to define a class that inherits all the methods and properties from another class.

**Parent class** is the class being inherited from, also called base class.

**Child class** is the class that inherits from another class, also called derived class.


```python
class Parent:
    
    parent_attri=120
    
    def __init__(self):
        print("display results of constructor")
    
    def parent_display(self):
        print("Hello I am Display method of parent class")
        
    def parent_attribute(self,attri):
        Parent.parent_attri=attri
        
    def parent_get_attribute(self):
        print("parent class attribute is :", Parent.parent_attri)
        
       
        
    
class Child(Parent):
            
    def __init__(self):
                print("child constructor")
                
    def child_display(self):
                print("Hi I am child method")
                    
```

```python
c=Child()
c.child_display()

# calling parent class's methods 
c.parent_display()
c.parent_attribute(100)  # pass the value of parent attribute 
c.parent_get_attribute() # display the parent attribute 
```

## Child class have multiple parent class

```python
class A:
    def __init__(self):
        print("Class A constructor")
        
    def method_A(self):
        print("Calling class A method")
        
class B:
    
    def __init__(self):
        print("Class B constructor")
        
    def method_B(self):
        print("Calling class B method")
# class C has multiple parent class A & B 
# so, class C can able to derive the both of the parent class's method...(method_A and method_B)
class C(A,B): 
    
    def __init__(self):
        print("Class C constructor")
        
    def method_C(self):
        print("Calling class C method")
                
```

```python
# creating instances of objects and child class calling(inherits) the parent class methods
c1=C()
c1.method_C()

#calling parent class A & B methods

c1.method_A()
c1.method_B()
```

## Overriding Methods

You can always override your parent class methods. One reason for overriding parent's methods is because you may want special or different functionality in your subclass.



```python
#Override means you have a same function name in Both parent and child class.
```

```python
class Parent:        
    def myMethod(self):
        print('Calling parent method')

class Child(Parent): 
    def myMethod(self):
        print('Calling child method')

c = Child()          # instance of child
c.myMethod()         # child calls overridden method
```

## Operator Overloading

Python operators work for **built-in classes**. But the **same operator** behaves **differently** with different types. 

For example, the + operator will perform arithmetic addition on two numbers, merge two lists, or concatenate two strings.

This feature in Python that allows the **same operator to have different meaning according to the context** is called **operator overloading.**

```python
'''
here if you want to add two Point objects data
but it will give and error 
'''
class Point:
    def __init__(self, x=0, y=0):
        self.x = x
        self.y = y


p1 = Point(1, 2)
p2 = Point(2, 3)
print(p1+p2)
```

```python

'''
here with the help of overide method __str__

this method(__str__) in our class that controls how the object gets printed.

'''
class Point:
    def __init__(self, x=0, y=0):
        self.x = x
        self.y = y

    def __str__(self):
        return "({0}, {1})".format(self.x, self.y)


p1 = Point(2, 3)
print(p1)
```

```python
str(p1) #built-in function calls the __str__ method
```

```python
format(p1) #built-in function calls the __str__ method
```

```python

class Point:
    def __init__(self, x=0, y=0):
        self.x = x
        self.y = y

    def __str__(self):
        return "({0},{1})".format(self.x, self.y)

    def __add__(self, other):
        x = self.x + other.x
        y = self.y + other.y
        return Point(x, y)


p1 = Point(1, 2)
p2 = Point(2, 3)

print(p1+p2)
```

### NOTE: when you use p1 + p2, Python calls p1.__add__(p2) which in turn is Point.__add__(p1,p2). After this, the addition operation is carried out the way we specified.

## Data Encapsulation

It is used to restrict access to methods and variables.

This can prevent the data to be modified by accident and is know as encapsulation.

```python
# Program to use Data Encapsulation
class st:
    def __init__(self):
        self.__show1()
    def show(self):
        print("Pooja Patel")
    def __show1(self):
        print("Hello ")
        
v=st()
v.show()
#v.__show1() # not accessible from object v bcz of private method
```

## Data Abstraction:

Sometimes we want that objects attributes may or may not be visible outside the class definition.

Here, we need to name attributes with double underscore prefix and those attributes then are not directly visible to the outsiders.


```python
# Program to use Data Hiding
class abc:
    __p=0
    def show(self):
        self.__p=self.__p+1
        print(self.__p)

v=abc()
v.show()
v.show()
v.show()
print(v.__p) # __p is hidden for outside class so error occurs
```
